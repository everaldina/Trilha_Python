```python
class Display(object):
    """Permite exibir representação HTML de vários objetos"""

    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

Com base nas dimensões de cada diamante se pode calcular o volume aproximado, considerando que eles como se fossem caixas. Adicione então uma nova coluna que mostre o preço por milímetro cubico de cada diamante. 


```python
#Fazer aqui o exercício 1
preco_por_mm3 = diamonds['price'] / (diamonds['x'] * diamonds['y'] * diamonds['z'])
diamonds['price/mm3'] = preco_por_mm3

Display('diamonds')
```

# Exercício 3

Utilizando os datasets anteriores identifique:
1. Quantos alunos de sexo masculino e feminino tem quem cada dataset. 


```python
masc_port = (port['sex'] == 'M').sum()
masc_math = (math['sex'] == 'M').sum()
fem_port = (port['sex'] == 'F').sum()
fem_math = (math['sex'] == 'F').sum()

genero_por_disciplina = pd.DataFrame({'Matemática': [masc_math, fem_math, masc_math + fem_math], 
                                      'Português': [masc_port, fem_port, masc_port + fem_port],
                                        }, index=['Masculino', 'Feminino', 'Total'])
genero_por_disciplina
```

2. Qual a média final dos alunos em cada disciplina.

```python
media_math = math['final_grade'].mean()
print(f'Média de notas de matemática: {media_math:.4}')

media_port = port['final_grade'].mean()
print(f'Média de notas de português: {media_port:.4}')
```

3. Qual a media final dos alunos cujos parentes (ao menos um) tem nível superior comparada à dos alunos em que nenhum dos pais tem esse tipo de formação.  

```python
# mascaras de ensino superior
mask_sup_port = (port['mother_education'] == 'higher education') | (port['father_education'] == 'higher education')
mask_sup_math = (math['mother_education'] == 'higher education') | (math['father_education'] == 'higher education')

# dados de ensino superior
sup_port = port[mask_sup_port]['final_grade']
not_sup_port = port[~mask_sup_port]['final_grade']

sup_math = math[mask_sup_math]['final_grade']
not_sup_math = math[~mask_sup_math]['final_grade']


#criando dataframe
dados_superior = pd.DataFrame({'Ensino Superior': [sup_port.max(), sup_port.min(), sup_port.mean(), sup_math.max(), sup_math.min(), sup_math.mean()], 
                               'Sem ensino superior': [not_sup_port.max(), not_sup_port.min(), not_sup_port.mean(), not_sup_math.max(), not_sup_math.min(), not_sup_math.mean()]},
                                 index=pd.MultiIndex.from_product([['Português', 'Matematica'], ['maximo', 'minimo', 'media']]))

dados_superior

```
