### __Exercício 1: Carregando e analisando um DataSet__

Dados de estações metereologicas por ano: [Instituto Nacional de Meteorologia](https://portal.inmet.gov.br/dadoshistoricos). 
1. Download de dados de 2003, 2013 e 2023 da estação Salvador.
2. Carregando e analizando dados das três estações em ``DataFrame``s separados. 

```python
import pandas as pd

# carregando os aquivos csv
estacao_ssa_2003 = pd.read_csv('datasets/estacao_Salvador_2003.csv', sep=";", header=8, encoding='latin-1')
estacao_ssa_2013 = pd.read_csv('datasets/estacao_Salvador_2013.csv', sep=",", header=8, encoding='utf-8')
estacao_ssa_2023 = pd.read_csv('datasets/estacao_Salvador_2023.csv', sep=";", header=8, encoding='latin-1')
```

```python
# estacao em 2023
estacao_ssa_2023.info()
estacao_ssa_2023.head()
```

* Limpando e unificando colunas para os três ``DataFrame``s.

```python
# removendo colunas extraas
estacao_ssa_2003.drop(columns=['Unnamed: 19'], inplace=True)
estacao_ssa_2023.drop(columns=['Unnamed: 19'], inplace=True)
```

```python
# unificando os dataframes
print("Colunas antes:")
colunas_2003 = estacao_ssa_2003.columns.values.tolist()
colunas_2013 = estacao_ssa_2013.columns.values.tolist()
colunas_2023 = estacao_ssa_2023.columns.values.tolist()

print(colunas_2003)
print(colunas_2013)
print(colunas_2023)

# renomeando colunas diferentes
print("\nColunas diferentes:")
for c_2003, c_2013, c_2023 in zip(colunas_2003, colunas_2013, colunas_2023):
    if c_2003 != c_2013 or c_2003 != c_2023:
        print(f"'{c_2003}' | '{c_2013}' | '{c_2023}'")
        estacao_ssa_2013.rename(columns={c_2013: c_2003}, inplace=True)
        estacao_ssa_2023.rename(columns={c_2023: c_2003}, inplace=True)


# imprimindo colunas depois
print("\nColunas depois:")
colunas_2003 = estacao_ssa_2003.columns.values.tolist()
colunas_2013 = estacao_ssa_2013.columns.values.tolist()
colunas_2023 = estacao_ssa_2023.columns.values.tolist()

print(colunas_2003)
print(colunas_2013)
print(colunas_2023)
```

### __Exercício 2: Juntando os datasets__

1. Constrindo um novo ``DataFrame`` utilizando um ``MultiIndex``.

```python
# convertendo a coluna data para o tipo datetime
estacao_ssa_2003['DATA (YYYY-MM-DD)'] = pd.to_datetime(estacao_ssa_2003['DATA (YYYY-MM-DD)'])
estacao_ssa_2013['DATA (YYYY-MM-DD)'] = pd.to_datetime(estacao_ssa_2013['DATA (YYYY-MM-DD)'])
estacao_ssa_2023['DATA (YYYY-MM-DD)'] = pd.to_datetime(estacao_ssa_2023['DATA (YYYY-MM-DD)'])


# pegando listas com os anos, meses e dias de cada dataframe
anos = pd.concat([estacao_ssa_2003['DATA (YYYY-MM-DD)'].dt.year,
                  estacao_ssa_2013['DATA (YYYY-MM-DD)'].dt.year,
                  estacao_ssa_2023['DATA (YYYY-MM-DD)'].dt.year])
anos = anos.values

meses = pd.concat([estacao_ssa_2003['DATA (YYYY-MM-DD)'].dt.month,
                   estacao_ssa_2013['DATA (YYYY-MM-DD)'].dt.month,
                   estacao_ssa_2023['DATA (YYYY-MM-DD)'].dt.month])
meses = meses.values

dias = pd.concat([estacao_ssa_2003['DATA (YYYY-MM-DD)'].dt.day,
                  estacao_ssa_2013['DATA (YYYY-MM-DD)'].dt.day,
                  estacao_ssa_2023['DATA (YYYY-MM-DD)'].dt.day])



# criando indice
index = pd.MultiIndex.from_arrays([anos, meses, dias], names=['ano', 'mes', 'dia'])
```

```python
# criando dataframa multiindexado por ano, mes e dia
estacoes_concat = pd.concat([estacao_ssa_2003, estacao_ssa_2013, estacao_ssa_2023])
estacao_ssa = estacoes_concat.drop(columns=['DATA (YYYY-MM-DD)', 'HORA (UTC)'])
estacao_ssa.set_index(index, inplace=True)
estacao_ssa
```

2. Construindo novo ``DataFrame`` utilizando ``MultiIndex`` como colunas contendo os parâmetros agrupados (precipitações, pressão, radiação, temperatura, umidade, velocidade do vento) 

```python
# criando multiindex para as colunas
nivel_1 = ['precipitacao'] + ['pressao']*3 + ['radiacao'] + ['temperatura']*6 + ['umidade']*3 + ['velocidade do vento']*3
nivel_2 = colunas_2003[2:]

# definindo multiindex para as colunas
index_col = pd.MultiIndex.from_arrays([nivel_1, nivel_2], names=['medicao', 'tipo'])

estacao_ssa.columns = index_col
estacao_ssa['HORA (UTC)'] = estacoes_concat['HORA (UTC)'].values # retornando a coluna hora
estacao_ssa
```

### __Exercício 3: Analisando os DataFrame__

* Preparando os dados
   

```python
import numpy as np
estacao_ssa.info()
estacao_ssa.head()
```

```python
# substituindo valores -9999 por NaN
estacao_ssa.replace(['-9999', -9999, -9999.0, '-9999.0'], value= np.nan, inplace=True)

estacao_ssa
```

```python
# tratando precipitacao
print(estacao_ssa['precipitacao']['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].unique())

# substiduindo valores , por .
estacao_ssa.loc[:, ('precipitacao', 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)')].replace(',', '.', regex=True, inplace=True)

# convertendo para float
estacao_ssa['precipitacao'] = estacao_ssa['precipitacao'].astype('float64')

# preenchendo valores NaN com valor anteriores
estacao_ssa.loc[:, ('precipitacao', 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)')].bfill(inplace=True)
```

```python
# tratando temperaturas
print(np.unique(estacao_ssa['temperatura'].values.astype('str')))
colunas = estacao_ssa['temperatura'].columns.values.tolist()

# substiduindo valores , por .
for coluna in colunas:
    estacao_ssa.loc[:, ('temperatura', coluna)].replace(',', '.', regex=True, inplace=True)

# convertendo para float
estacao_ssa['temperatura'] = estacao_ssa['temperatura'].astype('float64')

# preenchendo valores NaN com valor anteriores
for coluna in colunas:
    estacao_ssa.loc[:, ('temperatura', coluna)].bfill(inplace=True)
```

```python
# tratando hora

print(estacao_ssa['HORA (UTC)'].unique())
mask_utc = estacao_ssa['HORA (UTC)'].str.contains('UTC') 

# trantando dados utc
dados_utc = estacao_ssa.loc[mask_utc, 'HORA (UTC)']
dados_utc = dados_utc.replace(' UTC', '', regex=True) # removendo UTC
dados_utc = dados_utc.str[0:2] + ':00' # pegando apenas a hora

# colocando os dados tratados no dataframe
estacao_ssa.loc[mask_utc, 'HORA (UTC)'] = dados_utc.values
estacao_ssa['HORA (UTC)'].unique()
```

1. Analisando as colunas relacionadas a precipitações e temperaturas.

```python
# media de precipitacao por ano
precipitao_temp_ano = estacao_ssa[['precipitacao', 'temperatura']].groupby('ano')
precipitao_temp_ano.aggregate(['min', 'mean', 'max'])
```

2. Comparando dados para o mes de Janeiro. 


```python
precipitao_temp_janeiro = estacao_ssa[['precipitacao', 'temperatura']].groupby(['ano', 'mes']).aggregate(['min', 'mean', 'max'])
precipitao_temp_janeiro.query('mes == 1')
```

3. Comparando as informações sobre temperatura e precipitações medias em cada uma das estações dos três anos. 


```python
estacoes = {1: "verao", 2: "verao", 3:'verao', 
            4:'outono', 5:'outono', 6:'outono', 
            7:'inverno', 8:'inverno', 9:'inverno', 
            10:'primavera', 11:'primavera', 12:'primavera'}

precipitao_temp_estacao = estacao_ssa[['precipitacao', 'temperatura']].groupby(['mes']).mean().groupby(estacoes).mean()
precipitao_temp_estacao
```

4. Comparando informações sobre temperatura e precipitações medias para o dia 28 de setembro  em cada um dos intervalos do dia dos três anos.


```python
# vendo valores de hora
print(estacao_ssa['HORA (UTC)'].unique())

# criando diciionario para mapear as horas
horario = {'00:00': 'madrugada', '01:00': 'madrugada', '02:00': 'madrugada', '03:00': 'madrugada', '04:00': 'madrugada', '05:00': 'madrugada',
           '06:00': 'manha', '07:00': 'manha', '08:00': 'manha', '09:00': 'manha', '10:00': 'manha', '11:00': 'manha', '12:00': 'manha',
           '13:00': 'tarde', '14:00': 'tarde', '15:00': 'tarde', '16:00': 'tarde', '17:00': 'tarde', '18:00': 'tarde',
           '19:00': 'noite', '20:00': 'noite', '21:00': 'noite', '22:00': 'noite', '23:00': 'noite'}


# pegando dia 28 de setembro de cada ano
estacao_ssa_28_09 = estacao_ssa.query('mes == 9 and dia == 28')


# agrupando por hora
temp_hora = estacao_ssa_28_09.groupby('HORA (UTC)')['temperatura'].mean()
precipitacao_hora = estacao_ssa_28_09.groupby('HORA (UTC)')['precipitacao'].mean()

# concatenando os dataframes
precipitacao_temp_hora = pd.concat([precipitacao_hora, temp_hora], axis=1)
precipitacao_temp_hora
```

```python
# agrupando por horario
precipitacao_temp_hora.groupby(horario).mean()
```
